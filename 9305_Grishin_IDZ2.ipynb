2) Набор контрпримеров
<ul>
    <li>([])()[] - внутри круглой скобки стоит квадратная</li>
    <li>[()][()([])] - в круглой есть квадратные скобки</li>
    <li>[[[[[][]]]()[]]()()()][()()[()()()]([])[]] - внутри круглой скобки стоят квадратные</li>
    <li>[[[[   (   [][]]]()[]]()()()][()()[()()()]()[]]    ) - внутри круглой скобки стоит последовательность</li>
</ul>

5) Проверка на принадлежность грамматики классу LL(1)
<br>
<br>
Чтобы грамматика принадлежала классу LL(1),для каждого нетерминала, являющегося левой  частью нескольких правил: 
                         <$A$> ::= a 1 | a 2 | ... | an, необходимо, чтобы пересечение функций $FIRST(ai)$ и $FIRST(aj)$ было пусто для всех i =/= j. 
И чтобы Для каждого аннулирующего нетерминала <$A$>,такого что <$A$> ==>$\$$ необходимо, чтобы пересечение  множеств $FIRST(<$A$>)$ и $FOLLOW(<$A$>)$ было пустым. 
<br>
Полученная грамматика принадлежит классу LL(1), так как для нетеринала <$язык$ $псп$> $FIRST = \{[,( \}$, а  $FOLLOW = \{],\$\}$
<br>
То есть $FIRST \cap FOLLOW = \varnothing$ 
<br>Для <$ксп$> $FIRST = \{( \}$, а  $FOLLOW = \{),\$\}$,то есть $FIRST \cap FOLLOW = \varnothing$

9) Программа на языке Python:
В КОНЦЕ СТРОКИ НУЖНО ВСТАВЛЯТЬ СИМВОЛ "#".
ПРИМЕРЫ ПРАВИЛЬНЫХ ЗАПИСЕЙ:
<br>
[()()()()]
<br>
[()]()[()()][][()]
<br>
[()[()[(())]]]

ПРИМЕРЫ НЕПРАВИЛЬНЫХ ЗАПИСЕЙ:
<br>
([])()[] 
<br>
[()][()([])]

```python
import sys
i = -1
s = input()
errors = 0

def read():
    global i
    i += 1
    if i == len(s):
        return 0
    else:
        return s[i]

def error():
    global errors
    if(errors == 0):
        print("Ошибка!")
    errors = 1
    
def check_expression_psp(symb):
    if(symb == '['):
        symb = read()
        symb = check_expression_psp(symb)
        if(symb == ']'):
            symb = read()
        else:
            error()
        symb = check_expression_psp(symb)
    elif(symb == '('):
        symb = read()
        symb = check_expression_ksp(symb)
        if(symb == ')'):
            symb = read()
        else:
            error()
        symb = check_expression_psp(symb)
    return symb
def check_expression_ksp(symb):
    if(symb == '('):
        symb = read()
        symb = check_expression_ksp(symb)
        if(symb == ')'):
            symb = read()
        else:
            error()
        symb = check_expression_ksp(symb)
    return symb
flag = True

while(flag):
    symb = read()
    symb = check_expression_psp(symb)

    if(symb == '#'):
        print("Правильно!")
    print("Продолжить? 1 - Да; 0 - Нет")
    flag = bool(int(input()))
    if(flag == False):
        break
    i = -1
    s = input()
    errors = 0



```
